Теперь рассмотрим статистику по начличию устройств каждой из компаний отдельно:

```python
device_names = ['android', 'apple', 'web']
def visualize_company_devices(companies : list):
    fig, ax =plt.subplots(2,2, figsize = (12, 12))
    for i, company in enumerate(companies):
        curr_ax = ax[i//2][i%2]
        device_cols = [company + '_' + device for device in device_names]
        if (i >= 2):
            curr_ax.set_xlabel("company", fontsize = 23)
        else:
            curr_ax.set_xlabel("")
        if (i % 2 == 0):
            curr_ax.set_ylabel("numer", fontsize = 23)
            
        ax[i//2][i%2].set_title(company, fontsize = 23)
        sns.barplot(x = device_names, y = data[device_cols].sum(axis = 0), ax = curr_ax)
    plt.subplots_adjust(wspace=0.4, hspace = 0.3)
    fig.suptitle(f'num of devices in companies', fontsize = 30)
    plt.show()
```

```python
visualize_company_devices(companies)
```

##### Выводы про компанию:  
#### Black:
Число устройств отличается но не более чем в 2 раза, при этом продукцией apple(приложения ставят на эту марку) пользуются чаще всего (почти вдвое чаще чем android
#### Green:  
приложения установленные в этой компании в основном сосредоточена на android и apple, на ПК почти втрое меньше пользователей  
#### Blue:

#### Red:
apple очень не популярен, android'ом пользуются заметно чаще, в основном приложения ставятся на ПК

Сравнить компании по этому параметру достаточно сложно, поскольку отличие чуть менее чем на порядок, и дело может быть в неравномерном сборе данных или о том что некоторые компании гораздо популярнее

##### Рассмотрим теперь сколько пользователей находятся в Москве, которые зарегестрированы и сравним с числом зарегестрированных в целом

```python
def visualize_msk_users(companies : list):
    fig, ax = plt.subplots(1, len(companies), figsize = (12, 6))
    for i, company in enumerate(companies):
        col = 'has_' + company + '_id'
        has_id_mask = (data[col] != 0)
        msk_users = data[(data['is_msk']==1 & has_id_mask)]['is_msk'].sum() 
        all_users = data[has_id_mask][col].sum()
        
        ax[i].set_xlabel("company", fontsize = 25)
        if (i == 0):
            ax[i].set_ylabel("numer", fontsize = 25)
            
        ax[i].set_title(company, fontsize = 23)
        sns.barplot(x = ['all', 'msk'], y = [all_users, msk_users], ax = ax[i])
    plt.subplots_adjust(right = 1, wspace = 0.5)
    fig.suptitle(f'users in companies', fontsize = 30, y = 1.1)
    plt.show()
```

```python
visualize_msk_users(companies[:3])
```

##### Из графиков видно что:
- В Blue из москвы мало пользователей по сравнению с общим числом и другими компаниями  
- В Green наоборот очень много пользователей из Москвы(вообще говоря почти все) 
- В Black пользователей тоже достаточно много  
- сравнивать с Blue бесполезно из-за большого перевеса, отднако можно заметить что в Green сильно больше пользователей из Москвы чем в Black, хотя кол-во устройств компании Black заметно больше чем Green, значит что продукция Green используется чаще без приложений, причем заметно чаще в Москве чем в Black

Теперь попробуем посмотреть на корреляцию с помощбю heatmap.
Сначала посмотрим внутри каждой компании

```python
f, ax = plt.subplots(4, 1, figsize=(8, 32)) 
for i, company in enumerate(companies):
    curr_ax = ax[i]
    
    company_cols = [col for col in data.columns if (col.find(company) != -1)]
    
    labels = company_cols
    sns.heatmap(data[company_cols].corr(), ax = curr_ax, annot = True, robust = True, fmt=".2f")
    curr_ax.set_title(f'{company} correlations', fontsize = 23)
plt.subplots_adjust(hspace = 1)
plt.show()
```

Кроме очевидных корреляций явно видно:
- в компаниях Black и Green явно видна зависимость между наличием зарегестрированного аккаунта и gvm(для каждого поля), а в Blue зависимость хоть и есть но сильно меньше  
- внутри всех компаний нету видимых зависимостей между различными типами устройств(т.е. на чем установлено приложение)
