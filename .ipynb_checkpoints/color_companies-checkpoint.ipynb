Теперь рассмотрим статистику по начличию устройств каждой из компаний отдельно:

```python
device_names = ['android', 'apple', 'web']
def visualize_company_devices(companies : list):
    fig, ax =plt.subplots(2,2, figsize = (12, 12))
    for i, company in enumerate(companies):
        device_cols = [company + '_' + device for device in device_names]
        if (i >= 2):
            ax[i//2][i%2].set_xlabel("company")
        else:
            ax[i//2][i%2].set_xlabel("")
        if (i % 2 == 0):
            ax[i//2][i%2].set_ylabel("numer")
            
        ax[i//2][i%2].set_title(company, fontsize = 23)
        sns.barplot(x = device_names, y = data[device_cols].sum(axis = 0), ax = ax[i//2][i%2])
    plt.subplots_adjust(left=0, right = 1)
    fig.suptitle(f'num of devices in companies', fontsize = 30)
    plt.show()
```

```python
visualize_company_devices(companies)
```

##### Выводы про компанию:  
#### Black:
Число устройств отличается но не более чем в 2 раза, при этом продукцией apple(приложения ставят на эту марку) пользуются чаще всего (почти вдвое чаще чем android
#### Green:  
приложения установленные в этой компании в основном сосредоточена на android и apple, на ПК почти втрое меньше пользователей  
#### Blue:
В этой компании ПК самый популярный, при этом установленных приложений на apple почти вдвое меньше чем на андроид, а ПК на ~20% популярнее android в компании  
#### Red:
apple очень не популярен, android'ом пользуются заметно чаще, в основном приложения ставятся на ПК

Сравнить компании по этому параметру достаточно сложно, поскольку отличие чуть менее чем на порядок, и дело может быть в неравномерном сборе данных или о том что некоторые компании гораздо популярнее


#### Общие выводы про виды устройств:
apple наиболее популярен в Black и Green  
android умеренно популярен во всех компаниях  
ПК в целом популярен везде кроме Green  

- Из этого можно сказать что Red разрабатывает какой-то софт преимущественно для ПК, который слабо поддерживается apple и android - возможно антивирус или нечто подобное   
- Продукция Blue также разрабатывает приложения преимущественно для ПК, однако и на android и на apple достаточно часто приложения имеют место  
- Green же разрабатывает приложения преимущественно для android и apple, но иногда используют и на ПК, может быть какой-то непопулярный мессенжер  
- Blak же разрабытывает софт для всех видов устройст, который часто используется apple

##### Рассмотрим теперь сколько пользователей находятся в Москве, которые зарегестрированы и сравним с числом зарегестрированных в целом

```python
print(data['has_Blue_id'].unique())
```

```python
def visualize_msk_users(companies : list):
    fig, ax = plt.subplots(2,2, figsize = (12, 12))
    for i, company in enumerate(companies):
        has_id_mask = (data['has_' + company + '_id'] != 0)
        msk_users = data[(data['is_msk']==1 & has_id_mask)]['is_msk'].sum() 
        all_users = data[has_id_mask].sum()
        if (i >= 2):
            ax[i//2][i%2].set_xlabel("company")
        else:
            ax[i//2][i%2].set_xlabel("")
        if (i % 2 == 0):
            ax[i//2][i%2].set_ylabel("numer")
            
        ax[i//2][i%2].set_title(company, fontsize = 23)
        sns.barplot(x = ['all users', 'msk users'], y = pd.Data[all_users, msk_users], ax = ax[i//2][i%2])
    plt.subplots_adjust(left=0, right = 1)
    fig.suptitle(f'num of devices in companies', fontsize = 30)
    plt.show()
```

```python
visualize_msk_users(companies[:3])
```
